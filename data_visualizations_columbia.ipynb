```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure 'yr' is treated as integer
df2['yr'] = df2['yr'].astype(int)

# Replace blank/whitespace responses with "No Response Provided"
df2['Childinschool'] = df2['Childinschool'].replace(r'^\s*$', 'No Response Provided', regex=True)
response_translation = {
    'Sí': 'Yes',
    'No': 'No',
    'Prefiere no decir': 'Prefer not to Say',
    'dontkNow': "Don't Know",
    'No Response Provided': 'No Response Given'
}
df2['Childinschool'] = df2['Childinschool'].map(response_translation).fillna(df2['Childinschool'])

# Drop duplicates based on unique id + year + quarter
df_unique = df2.drop_duplicates(subset=['id', 'yr', 'quarter'])

# Group and calculate counts
grouped = df_unique.groupby(['yr', 'Childinschool'])['id'].nunique().reset_index(name='count')

# Calculate percentage within each year group
grouped['percentage'] = grouped.groupby('yr')['count'].transform(lambda x: (x / x.sum()) * 100)

# Sort responses within each year
grouped = grouped.sort_values(['yr', 'percentage'], ascending=[True, False])

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped, x='yr', y='percentage', hue='Childinschool', palette='Set2')

# Add percentage labels
for container in plt.gca().containers:
    for bar in container:
        height = bar.get_height()
        if height > 0:
            plt.gca().text(
                bar.get_x() + bar.get_width() / 2,
                height + 0.5,
                f'{height:.1f}%',
                ha='center',
                va='bottom',
                fontsize=9
            )

# Formatting
plt.title('Child in School Response Distribution by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.legend(title='Response', fontsize=10, title_fontsize=11)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('columbia_SchoolAttendance.png')
plt.show()
```
